```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis, y_axis, 'o', marker='x', color="red", label='Chicago')

plt.xlabel('months')
plt.ylabel('Total Fare ($)')
plt.grid()
# Add the legend.
plt.legend()
```

```python
# For Bubble Chart. 
#Note. After running this cell,the marker changes in size and color based on the value of the fare in USD
plt.scatter(x_axis, y_axis)
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# We can adjust the size by multiplying the data in the y-axis by any value. Let's multiply each data point in the y-axis by 3 and see what happens.
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger, edgecolors="black")
plt.title("Total Fare ($) v Months")
```

```python
#we can use Python's list comprehension technique inside the plt.scatter() function
#to multiply each data point in the y-axis list, we would do the following in the plt.scatter() function:
#without using the for loop function above
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis], edgecolors="black")
plt.title("Total Fare ($) v Months")
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, shadow=True, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
